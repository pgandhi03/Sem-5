```python
#read stockcode nd price
df[["StockCode","UnitPrice"]].head()
```

```python
price = df["UnitPrice"]
```

```python
print("Minimum price:",price.min())
print("Maximum price:",price.max())
print("Median of price:",price.median())
print("Median of price:",st.median(price))
print("Mean of price:",price.mean())
print("Mean of price:",st.mean(price))
print("Count of price:",price.count())
print("Sum of price:",price.sum())
print("Standard Deviation:",price.std())
print("Standard Deviation:",st.stdev(price))
print("Variance:",price.var())
print("Variance:",st.variance(price))
```
