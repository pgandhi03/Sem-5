```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Advertising.csv",index_col=0)
```

```python
df.head()
```

```python
#missing data
df.isna()
```

```python
#categorize
conditions = [(df["Sales"] >= 15),
             (df["Sales"] < 15) & (df["Sales"] >= 5),
             (df["Sales"] < 5)]
values = ["High", "Medium", "Low"]
sale_type = pd.DataFrame({"Sales Type":np.select(conditions,values)})
```

```python
sale_type.head()
```

```python
#Concatenation
df["Sales Type"] = np.select(conditions,values)
```

```python
df.head()
```

```python
#Groupby
df.groupby('Sales Type').agg(['sum',"mean"])
```

```python
#data modeling
```

```python
df.shape[0]
```

```python
df.head()
```

```python
#creating nan values
for i in range(6,df.shape[0]+1,6):
    df["TV"][i] = None
for i in range(10,df.shape[0]+1,10):
    df["Newspaper"][i] = None
```

```python
#printing rows without NaN value
df[np.all(df.notnull(), axis = 1)]
```

```python
#Filling Nan values with 0
df["TV"].fillna(0, inplace = True)
```

```python
df.head(10)
```

```python
#Filling Nan values with average of just above and just below value
df["Newspaper"].interpolate(inplace = True)
```

```python
df.head(12)
```

```python
#creating NaN again
for i in range(10,df.shape[0]+1,10):
    df["Newspaper"][i] = None
```

```python
#Filling NaN values with mean of the column 
df["Newspaper"].fillna(df["Newspaper"].mean(), inplace = True)
```

```python
df.head(10)
```

```python
#Normalization
```

```python
import sklearn.preprocessing as sp
```

```python
for i in df.columns[:3]:
    df[i] = sp.minmax_scale( df[i] )
```

```python
df.columns[:3]
```

```python
#splitting
from sklearn.model_selection import train_test_split
```

```python
X = df[df.columns[:3]]
y = df["Sales"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.4)
```

```python
#training
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
help(LinearRegression)
```

```python
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
model.score(X_test, y_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
print("MSE: ",mean_squared_error(y_test,y_pred))
print("R2_score:  ",r2_score(y_test, y_pred))
```
