```python
#Affine Cipher
def inverse(a,m):
    x = 1
    y = 0
    m0 = m
    while(a>1):
        q = a//m
        t = m
        m = a%m
        a = t
        t = y
        
        y = x - q * y
        x = t
    if(x<0):
        x += m0
    return x
print("Enter keys such that a and m(=26) are coprime")
m = 26
a = int(input("Enter a: "))
b = int(input("Enter b: "))
choice = int(input("Enter 1 for encryption and 2 for decryption: "))
if(choice == 1):
    msg = input("Enter plaintext: ")
    msg = msg.upper()
    en = ""
    for i in msg:
        if(i==" "):
            en += " "
        else:
            p = ord(i)-65 #number of that character
            c = ( (a*p) + b ) % m
            en += chr(65+c)

    print("Encrypted text: ",en)
elif(choice == 2):
    msg = input("Enter encrypted text: ")
    msg = msg.upper()
    ain = inverse(a,m)
    de = ""
    for i in msg:
        if(i==" "):
            de += " "
        else:
            c = ord(i)-65 #number of that character
            p = ( ain * (c - b) ) % m
            de += chr(65+p)
    print("Decrypted text: ", de)
        
```
